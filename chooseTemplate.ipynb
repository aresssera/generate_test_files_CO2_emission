```python
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET
import csv
```

```python
fileName = 'input/allPossibleCombinations.csv'
df = pd.read_csv(fileName)
```

```python
df['ClassOfHybridVehicleCode'].unique()
```

```python
df = df.astype({"#FuelCode1": str}, errors='raise')
```

```python
templates = {'NOVC-HEV': ['templates/Temp_NOVC_Monofuel.xml',
                          'templates/Temp_NOVC_Bifuel.xml',
                          'templates/Temp_NOVC_Trifuel.xml'],
             'OVC-HEV': ['templates/Temp_OVC_Monofuel.xml',
                         'templates/Temp_OVC_Bifuel.xml',
                         'templates/Temp_OVC_Trifuel.xml'],
             'NOVC-FCHV': ['templates/Temp_NOVC-FCHV.xml'],
             'OVC-FCHV':  ['templates/Temp_OVC-FCHV.xml'],
             'Empty': ['templates/Temp_NoHybrid_Monofuel.xml',
                       'templates/Temp_NoHybrid_Bifuel.xml',
                       'templates/Temp_NoHybrid_Trifuel.xml']}
```

```python
fuelCodes = ['#FuelCode1', '#FuelCode2', '#FuelCode3']

def chooseTemplate(row):

  fuel = 0

  for fuelCode in fuelCodes:

    currFuel = row[fuelCode]
    if currFuel not in ['-', '0']:
      fuel += 1


  chvc = row['ClassOfHybridVehicleCode']

  if 'OVC' in chvc:

    temps = templates[chvc]

    if len(temps) > 1:
      return temps[fuel-1]
    else:
      return temps[0]

  else:
    fuelcode1 = row['#FuelCode1']
    fuelcode4 = row['FuelCode4']

    if fuelcode1 == '0':
      return 'templates/Temp_REINELEKTRISCH_E.xml'
    elif fuelcode1 == '50' and fuelcode4 == '-':
      return 'templates/Temp_Temp_WASSERSTOFF_W.xml'
    else:
      return templates['Empty'][fuel-1]

```

```python
df['Template'] = df.apply(lambda row: chooseTemplate(row), axis=1)
```
